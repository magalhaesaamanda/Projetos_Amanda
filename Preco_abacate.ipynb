# Tratamento de dados

## Tratamento nomes das cidades

```python
regions = [i for i in abacate.region.unique()]
```

```python
regions_altera = regions[:]
caracters = ['/', " ", '.']

for x in range(len(regions)):
  for y in caracters:
    regions_altera[x] = regions_altera[x].replace(y, "")
```

```python
def subs_nome(dataframe, n1, n2):
  dataframe.loc[dataframe['region']== n1, 'region'] = n2
```

```python
for i in range(len(regions)):
  subs_nome(abacate, regions[i], regions_altera[i])
```

## Criando coluna com outros tipos de abacate

```python
abacate['ab_outr'] = (abacate.vol_total -  (abacate.ab_peq + abacate.ab_grd + abacate.ab_xgrd))
```
