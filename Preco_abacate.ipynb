# Tratamento de dados

## Tratamento nomes das cidades

```python
regions = [i for i in abacate.region.unique()]
```

```python
regions_altera = regions[:]
caracters = ['/', " ", '.']

for x in range(len(regions)):
  for y in caracters:
    regions_altera[x] = regions_altera[x].replace(y, "")
```

```python
def subs_nome(dataframe, n1, n2):
  dataframe.loc[dataframe['region']== n1, 'region'] = n2
```

```python
for i in range(len(regions)):
  subs_nome(abacate, regions[i], regions_altera[i])
```

```python
abacate['region'].unique()
```

## Removendo linhas duplicadas

```python
# Removendo as linhas duplicadas

abacate = abacate.drop_duplicates()
```

## Trocando nome das colunas


```python
abacate.head()
```

```python
abacate.rename(columns = {'Date': 'data', 'AveragePrice': 'preco_medio', 'Total Volume': 'vol_total', '4046': 'ab_peq', '4225': 'ab_grd',
                          '4770': 'ab_xgrd', 'Total Bags': 'total_sacolas', 'Small Bags': 'sac_peq','Large Bags': "sac_grd",'XLarge Bags': "sac_xgrd",
                          'type': 'tipo', 'year': 'ano', 'region': 'regiao'}, inplace = True)
```

```python
abacate.head()
```

## Criando coluna com outros tipos de abacate

```python
abacate['ab_outr'] = (abacate.vol_total -  (abacate.ab_peq + abacate.ab_grd + abacate.ab_xgrd))
```

```python
abacate.head()
```

## Criando coluna com o tipo de abacate de forma numérica



```python
# Adotando os valores 0 para conventional e 1 para organic

abacate.loc[abacate.tipo == 'conventional', 'tipo_abacate'] = int(0)
abacate.loc[abacate.tipo == 'organic', 'tipo_abacate'] = int(1)
```

```python
abacate.head()
```

## Ordenando colunas e resetando o index

```python
# Criando a ordem para as colunas do dataframe

ordem_colunas = ['data', 'ano', 'regiao', 'sac_peq', 'sac_grd', 'sac_xgrd', 'total_sacolas', 'tipo', 'tipo_abacate', 'ab_peq', 'ab_grd', 'ab_xgrd', 'ab_outr', 'vol_total', 'preco_medio']
abacate = abacate[ordem_colunas].reset_index(drop=True)
```
